# PreProccessing Data

```python
doors_mapping = {'04-May':'4','02-Mar':'2','>5':'5'}
```

```python
df['Doors'] = df['Doors'].map(doors_mapping)
```

```python
def remove_km(x):
    return x[:-2]
df['Mileage'] = df['Mileage'].apply(remove_km)
df['Mileage'] = pd.to_numeric(df['Mileage'])
```

```python
df['Levy'] = df['Levy'].replace({'-':'0'})
df['Levy'] = pd.to_numeric(df['Levy'])
```

```python
df['Levy'] = df['Levy'].replace({0:np.round(df['Levy'].mean())})
```

```python
dff = df.copy()
```

```python
outlier_price = dff['Price'].sort_values(ascending=False).head(3).index
dff.drop(outlier_price,inplace=True)
```

```python
sns.scatterplot(data=dff,x='Levy',y='Price')
```

```python
l = dff['Mileage'].quantile(0.25)
h = dff['Mileage'].quantile(0.75)
iqr = h-l
loww = l - 1.5*iqr
highh = h + 1.5*iqr
print(loww,highh)
```

```python
x=dff[(dff['Price']>dff['Levy']) & (dff['Price'] > 1000) & (dff['Price']<150000) & (dff['Levy']<4000) & (dff['Mileage'] < 367588)]
x.describe()
```

```python
sns.scatterplot(data=x,x='Levy',y='Price')
```

```python
for label,content in x.items():
    if pd.api.types.is_string_dtype(content):
        x[label] = pd.Categorical(content).codes 
```

```python
x.drop('ID',axis=1,inplace=True)
```

```python
x
```

```python
features = x.drop('Price',axis=1)
target = x['Price']
target_log = np.log(target)
X_train, X_test, y_train, y_test = train_test_split(features,target,test_size=0.2, random_state=134)
#X_train, X_test, y_train, y_test = train_test_split(features,target_log,test_size=0.2, random_state=134)
```

```python
train_leater = X_train['Leather interior'].reset_index(drop=True)
X_train.drop('Leather interior',axis=1,inplace=True)

test_leather = X_test['Leather interior'].reset_index(drop=True)
X_test.drop('Leather interior',axis=1,inplace=True)
```

```python
scaller = StandardScaler()
xt_scalled = pd.DataFrame(data=scaller.fit_transform(X_train),columns=X_train.columns)
xs_scalled = pd.DataFrame(data=scaller.transform(X_test),columns=X_test.columns)
```

```python
xt_scalled_new = pd.concat([xt_scalled,train_leater],axis=1)
xs_scalled_new = pd.concat([xs_scalled,test_leather],axis=1)
```

```python
models = {
    'LinearRegression':LinearRegression(),
    'Ridge':Ridge(),
    'DecisionTree':DecisionTreeRegressor(),
    'Adaboost':AdaBoostRegressor(),
    'ExtraTree':ExtraTreesRegressor(),
    'RandomForest':RandomForestRegressor(),
    'GradientBoosting':GradientBoostingRegressor(),
    'Kneighbor':KNeighborsRegressor(),
    'SVR':SVR(),
    'XGB':XGBRegressor()
}
```

```python
result = pd.DataFrame(columns=['R2','MSE',"MAE",'RMSE'])
def regression(x_train,x_test,y_train,y_test):
    for label,content in models.items():
        content.fit(x_train,y_train)
        prediction=content.predict(x_test)
        r2 = r2_score(y_test,prediction)
        mae = mean_absolute_error(y_test,prediction)
        scores = cross_val_score(content, x_train, y_train, scoring='neg_mean_squared_error', cv=5)
        rmse = (np.sqrt(-scores)).mean()
        result.loc[label,:]=[r2,(scores.mean())*-1,mae,rmse]
        scores = cross_val_score(content, x_train, y_train, scoring='neg_mean_squared_error', cv=5)
    return result
    
```

```python
regression(xt_scalled_new,xs_scalled_new,y_train,y_test)
```

```python
regression(xt_scalled_new,xs_scalled_new,y_train,y_test)
```
