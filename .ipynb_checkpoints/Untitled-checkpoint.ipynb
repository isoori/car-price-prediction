# PreProccessing Data

```python
doors_mapping = {'04-May':'4','02-Mar':'2','>5':'5'}
```

```python
df['Doors'] = df['Doors'].map(doors_mapping)
```

```python
def remove_km(x):
    return x[:-2]
df['Mileage'] = df['Mileage'].apply(remove_km)
df['Mileage'] = pd.to_numeric(df['Mileage'])
```

```python
df['Levy'] = df['Levy'].replace({'-':'0'})
df['Levy'] = pd.to_numeric(df['Levy'])
```
